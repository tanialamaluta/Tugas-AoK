PCA akan cukup efektif jika antar variabel asal memiliki korelasi yang cukup tinggi. Tujuan PCA adalah mendapatkan vektor yang menyatakan kembali dataset noisy yang sangat berarti. PCA berasumsi bahwa transformasi dapat dilakukan secara linier. Di sisi lain, PCA merupakan suatu transformasi ortogonal dari sistem koordinat dimana data diuraikan.
